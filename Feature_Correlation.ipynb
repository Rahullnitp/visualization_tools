# correlation 
  analysis of all the feature simultaneously of how they are correlated to each other.

```
import os
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```
data=pd.DataFrame(np.random.rand(100,8),columns=list('ABCDEFGH'))
```

```
data.head()
```

##a.Correlation matrix 
shows the relation between different features based on the colour of the corresponding square.

```
#correlation matrix(heatmap style)
corrmat = data.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

##b.correlation matrix (zoomed heatmap style)
it return the relation between top specified numbers of correlated features.

```
k = 6#number of variables for heatmap
cols = corrmat.nlargest(k, 'A')['A'].index
cm = np.corrcoef(data[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

##c.Pairplot
Draw scatterplots for joint relationships and histograms for univariate distributions:

```
#scatterplot
sns.set()
cols = ['A','B','C','D','E']
sns.pairplot(data[cols], size = 2.5)
plt.show();
```
